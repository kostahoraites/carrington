```python
# PCA example     https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

import numpy as np
from sklearn.decomposition import PCA
X = np.array([[-1, -1, 4], [-2, -1, 5], [-3, -2, 0], [1, 1, -1], [2, 1, 4], [3, 2, 1]])
pca = PCA()   # n_components = X
pca.fit(X)
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
print(pca.components_)
X.shape
```
